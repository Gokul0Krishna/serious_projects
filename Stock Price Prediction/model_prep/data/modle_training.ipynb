```python
import pandas as pd
import optuna
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
! unzip /content/cleaned_data.zip
```

```python
dft = pd.read_csv('/content/cleaned_data.csv')
dft["lag_1"] = dft["Close"].shift(1)
dft["lag_2"] = dft["Close"].shift(2)
dft["lag_3"] = dft["Close"].shift(3)
dft["rolling_7"] = dft["Close"].rolling(7).mean()
dft["rolling_14"] = dft["Close"].rolling(14).mean()
dft["rolling_std_7"] = dft["Close"].rolling(7).std()
df = dft.dropna()
tree_features = [
    "Open","High","Low","Close","Volume",
    "rsi","ema_10","ema_50","macd",
    "bb_high","bb_low",
    "lag_1","lag_2","lag_3",
    "rolling_7","rolling_14","rolling_std_7"
]

X_tree = dft[tree_features].values
y_tree = dft["target"].values
X_train,X_test,ytrain,ytest = train_test_split(X_tree, y_tree ,test_size=0.2,random_state=42)
X_test,X_val,ytest,yval = train_test_split(X_tree, y_tree ,test_size=0.5,random_state=42)
```

```python
model = xgb.XGBRegressor(objective='reg:squarederror',
                         n_estimators=900, random_state=42)
model.fit(X_train, ytrain)
preds = model.predict(X_val)
mse = mean_squared_error(yval, preds)
```
