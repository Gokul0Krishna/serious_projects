Data explaored now the actual work to be done

```python
dfz.rename(columns={
    'variable':'d',
    "value":'demand'
},inplace=True
)
```

```python
dfz2= dfz.merge(df1,on="d",how="left")
```

```python
dfz2[:5]
```

```python
dfz2['date'] = pd.to_datetime(dfz2['date'])
dfz2["day_of_week"] = dfz2["date"].dt.weekday
dfz2["week_of_year"] = dfz2["date"].dt.isocalendar().week.astype(int)
dfz2["is_weekend"] = dfz2["day_of_week"].isin([5,6]).astype(int)
```

```python
dfz2[:5]
```

```python
dfz3 = dfz2.merge(df3,
                  on=["store_id", "item_id", "wm_yr_wk"],
                how="left"
                  )

dfz3[:5]
```

```python
dfz3["sell_price"] = (
    dfz3
    .groupby(["item_id", "store_id"])["sell_price"]
    .ffill()
)
```

```python
dfz3[:5]
```

```python
ts = dfz3[
    (dfz3["item_id"] == "HOBBIES_1_001") &
    (dfz3["store_id"] == "CA_1")
].sort_values("date")

ts[["date", "demand", "sell_price"]].head(10)
```
